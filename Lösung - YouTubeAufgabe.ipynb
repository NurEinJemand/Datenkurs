```python
import pytube
from pytube import Search
import time
from datetime import datetime
import numpy as np
import pandas as pd
```

```python
liste = []
fertigeListe = []
titelListe = []

s = Search('Minecraft')
ergebnis = s.results

for i in range(len(ergebnis)):
    string = ergebnis[i]
    liste = liste + [[string.title, string.length, string.views, string.video_id]]
    titelListe = titelListe + [string.title]

liste = np.array(liste)
DF = pd.DataFrame(liste)
DF.to_csv("Suchanfrage.csv", header=['Titel', 'Länge in Sekunden', 'Aufrufe', 'ID'])
```

```python
x = []
y1 = []
y2 = []

for i in range(180):

    if i > 0:
        time.sleep(10)

    yt1 = pytube.YouTube(url='https://www.youtube.com/watch?v=5m2uB9XwjBg')
    yt2 = pytube.YouTube(url='https://www.youtube.com/watch?v=i4NI94NO3eo')

    date = datetime.now()
    i = i + 1
  
    y1 = y1 + [yt1.views]
    y2 = y2 + [yt2.views]
    x = x + [time.time()]

yArrayF = [x] + [y1] + [y2]
yArrayF = np.array(yArrayF)
yArrayF = yArrayF.transpose()

DF = pd.DataFrame(yArrayF)

DF.to_csv("DatenYTVideos.csv", header=['Zeit', yt1.title, yt2.title]) 
```
